## !!! GAME OF SETS !!!

```python
from random import randint
from IPython.display import clear_output
from time import sleep
      
def same(a,b):
    while a == b:
        b = input("Sorry, it has already been taken.. please select another:\n").title()
    return b
def taken(a,pos):
    while True:
        if pos in a.keys():
            pos = randint(1,52)
        else:
            return pos
def cards(cards):
    for c in ["H","S","D","C"]:
        for i in range(1,10):
            pos = randint(1,52)
            pos = taken(cards,pos)
            cards[pos] = c+str(i)+" "
        pos = randint(1,52)
        pos = taken(cards,pos)
        cards[pos] = c+"10"
        for j in ["J","Q","K"]:
            pos = randint(1,52)
            pos = taken(cards,pos)
            cards[pos] = c+j+" "
def hide(a):
    for i in range(1,10):
        a.append(" "+str(i)+" ")
    for i in range(10,53):
        a.append(" "+str(i))                      
def match(a,b,card):
    if card[a][1] == card[b][1] and card[a][2] == card[b][2]:
        return True
    else:
        return False
def sets(s):
    print("┌───┬───┬───┬───┬───┬───┬───┬───┬───┬───┐")
    print("│"+str(s[0])+"│"+str(s[1])+"│"+str(s[2])+"│"+str(s[3])+"│"+str(s[4])+"│"+str(s[5])+"│"+str(s[6])+"│"+str(s[7])+"│"+str(s[8])+"│"+str(s[9])+"│")
    print("├───┼───┼───┼───┼───┼───┼───┼───┼───┼───┤")
    print("│"+str(s[10])+"│"+str(s[11])+"│"+str(s[12])+"│"+str(s[13])+"│"+str(s[14])+"│"+str(s[15])+"│"+str(s[16])+"│"+str(s[17])+"│"+str(s[18])+"│"+str(s[19])+"│")
    print("├───┼───┼───┼───┼───┼───┼───┼───┼───┼───┤")
    print("│"+str(s[20])+"│"+str(s[21])+"│"+str(s[22])+"│"+str(s[23])+"│"+str(s[24])+"│"+str(s[25])+"│"+str(s[26])+"│"+str(s[27])+"│"+str(s[28])+"│"+str(s[29])+"│")
    print("├───┼───┼───┼───┼───┼───┼───┼───┼───┼───┤")
    print("│"+str(s[30])+"│"+str(s[31])+"│"+str(s[32])+"│"+str(s[33])+"│"+str(s[34])+"│"+str(s[35])+"│"+str(s[36])+"│"+str(s[37])+"│"+str(s[38])+"│"+str(s[39])+"│")
    print("├───┼───┼───┼───┼───┼───┼───┼───┼───┼───┤")
    print("│"+str(s[40])+"│"+str(s[41])+"│"+str(s[42])+"│"+str(s[43])+"│"+str(s[44])+"│"+str(s[45])+"│"+str(s[46])+"│"+str(s[47])+"│"+str(s[48])+"│"+str(s[49])+"│")
    print("├───┼───┼───┴───┴───┴───┴───┴───┴───┴───┘")
    print("│"+str(s[50])+"│"+str(s[51])+"│")
    print("└───┴───┘")
        
player1 = input("Player1 enter your name:").title()
mark1 = input(player1 + " enter your mark:").upper()
player2 = input("Player2 enter your name:").title()
player2 = same(player1,player2)
mark2 = input(player2 + " enter your mark:").upper()
mark2 = same(mark1,mark2)
mark1 = " " + mark1 + " "
mark2 = " " + mark2 + " "
clear_output()
cardss = {}
cards(cardss)
hidden = []
hide(hidden)
counter = 1
while len(cardss) > 1:
    sets(hidden)
    if counter%2 != 0:
        player = player1
        mark = mark1
    else:
        player = player2
        mark = mark2
                    
    flip = int(input(player + " please select a card:"))
    while flip not in cardss.keys():
        flip = int(input(player + " its taken.. please select a another card:"))
    temp1 = hidden[flip-1]
    hidden[flip-1] = cardss[flip]
    clear_output()
    sets(hidden)
    flip2 = int(input(player + " please select 2nd card:"))
    while flip2 not in cardss.keys() or flip == flip2:
        flip2 = int(input(player + " kindly select a another card:"))
    if match(flip,flip2,cardss):
        clear_output()
        hidden[flip2-1] = cardss[flip2]
        sets(hidden)
        print("\n*********************\n!!! CARDS MATCHED !!!\n*********************")
        sleep(3)
        clear_output()
        hidden[flip-1] = mark
        hidden[flip2-1] = mark
        cardss.pop(flip)
        cardss.pop(flip2)
        counter -= 1
    else:
        clear_output()
        temp2 = hidden[flip2-1]
        hidden[flip2-1] = cardss[flip2]
        sets(hidden)
        print("\n       Sorry\nCards did not match")
        sleep(3)
        clear_output()
        hidden[flip-1] = temp1
        hidden[flip2-1] = temp2
    counter += 1
print(player1+"'s total sets = ",hidden.count(mark1)//2)
print(player2+"'s total sets = ",hidden.count(mark2)//2)
if hidden.count(mark1) > hidden.count(mark2):
    print(player1, "wins !!!")
elif hidden.count(mark1) < hidden.count(mark2):
    print(player2, "wins !!!")
else:
    print("Its a tie !!!")
        
```
